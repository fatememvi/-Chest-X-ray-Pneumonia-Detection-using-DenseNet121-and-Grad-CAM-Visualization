**Handling Class Imbalance with Weighted Sampling and Loss**

```python
# Override val transforms
val_dataset.dataset.transform = val_test_transforms


# ----- Weighted Sampler for imbalanced classes -----
# Get class counts
class_counts = [0]*len(train_dataset.dataset.classes)
for _, label in train_dataset:
    class_counts[label] += 1

# Compute weights: inverse of frequency
class_weights = [sum(class_counts)/c for c in class_counts]

# Assign weight to each sample
samples_weights = [0]*len(train_dataset)
for idx, (_, label) in enumerate(train_dataset):
    samples_weights[idx] = class_weights[label]

samples_weights = torch.tensor(samples_weights, dtype=torch.float)
sampler = WeightedRandomSampler(samples_weights, num_samples=len(samples_weights), replacement=True)

# ----- Weighted Loss -----
weights = torch.tensor(class_weights, dtype=torch.float)
criterion = nn.CrossEntropyLoss(weight=weights)
```

**Model Training Loop with Early Stopping and Scheduler**

```python
from tqdm import tqdm
```

```python
# ----- Hyperparameters -----
epochs = 20          # Number of epochs
batch_size = 32      # Batch size
learning_rate = 1e-4
patience = 5         # Early stopping patience

# ----- Device -----
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# ----- Model (DenseNet121) -----
from torchvision import models
import torch.nn as nn

def build_densenet121(num_classes=2, freeze_features=False):
    model = models.densenet121(weights='IMAGENET1K_V1')
    model.features.conv0 = nn.Conv2d(1, 64, kernel_size=7, stride=2, padding=3, bias=False)

    if freeze_features:
        for param in model.features.parameters():
            param.requires_grad = False

    in_features = model.classifier.in_features
    model.classifier = nn.Linear(in_features, num_classes)
    return model

model = build_densenet121(num_classes=2, freeze_features=False).to(device)
print("✅ DenseNet121 model loaded on:", device)

# ----- Weighted Loss for Imbalanced Classes -----
weights = torch.tensor(class_weights, dtype=torch.float).to(device)
criterion = nn.CrossEntropyLoss(weight=weights)

# ----- Optimizer -----
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate, weight_decay=1e-5)

# ----- Learning Rate Scheduler -----
scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=epochs)

# ----- Tracking Lists -----
train_loss, val_loss = [], []
train_accuracy, val_accuracy = [], []

# ----- Early Stopping -----
best_val_loss = float('inf')
trigger_times = 0

# ----- Directory for Checkpoints -----
timestamp = datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
checkpoint_dir = f"checkpoints_{timestamp}"
os.makedirs(checkpoint_dir, exist_ok=True)

# ----- Training Loop -----
print(f"🔧 Training started for {epochs} epochs on device: {device}\n")

for epoch in range(epochs):
    print(f"📘 Epoch {epoch+1}/{epochs}")
    model.train()
    correct, train_running_loss = 0, 0.0

    # 🔄 TRAIN LOOP
    for data, labels in tqdm(train_loader, desc=f"Training Epoch {epoch+1}", colour="cyan"):
        data, labels = data.to(device), labels.to(device)
        optimizer.zero_grad()

        outputs = model(data)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        train_running_loss += loss.item()
        _, preds = torch.max(outputs, 1)
        correct += (preds == labels).sum().item()

    epoch_train_loss = train_running_loss / len(train_loader)
    epoch_train_acc = 100 * correct / len(train_dataset)
    train_loss.append(epoch_train_loss)
    train_accuracy.append(epoch_train_acc)

    # 📊 VALIDATION LOOP
    model.eval()
    val_correct, val_running_loss = 0, 0.0
    with torch.no_grad():
        for val_data, val_labels in tqdm(val_loader, desc="Validating", colour="magenta"):
            val_data, val_labels = val_data.to(device), val_labels.to(device)
            val_outputs = model(val_data)
            val_loss_value = criterion(val_outputs, val_labels)
            val_running_loss += val_loss_value.item()
            _, val_preds = torch.max(val_outputs, 1)
            val_correct += (val_preds == val_labels).sum().item()

    epoch_val_loss = val_running_loss / len(val_loader)
    epoch_val_acc = 100 * val_correct / len(val_dataset)
    val_loss.append(epoch_val_loss)
    val_accuracy.append(epoch_val_acc)

    # 🧠 Report
    print(f"🔹 Training Loss: {epoch_train_loss:.4f} | Acc: {epoch_train_acc:.2f}%")
    print(f"🔸 Validation Loss: {epoch_val_loss:.4f} | Acc: {epoch_val_acc:.2f}%")

    # 💾 Save Best Model + Early Stopping
    if epoch_val_loss < best_val_loss:
        best_val_loss = epoch_val_loss
        trigger_times = 0
        best_model_path = os.path.join(checkpoint_dir, "best_model.pth")
        torch.save(model.state_dict(), best_model_path)
        print(f"✅ Best model saved at {best_model_path} (Val Loss: {best_val_loss:.4f})")
    else:
        trigger_times += 1
        if trigger_times >= patience:
            print(f"⏹ Early stopping triggered after {epoch+1} epochs")
            break

    # 🔄 Scheduler step
    scheduler.step()
    print(f"📉 LR updated to: {scheduler.get_last_lr()[0]:.6f}\n")

print("\n🎯 Training complete!")
print(f"Best Validation Loss: {best_val_loss:.4f}")
```

**Grad-CAM Visualization with Lung Masking for Chest X-ray Classification**

```python
import torch.optim as optim
from torch.utils.data import DataLoader
from torchvision import datasets, transforms, models


from pytorch_grad_cam import GradCAMPlusPlus
from pytorch_grad_cam.utils.model_targets import ClassifierOutputTarget
from pytorch_grad_cam.utils.image import show_cam_on_image

# ======================
# 1️⃣ Pretrained U-Net for Lung Segmentation (demo)
# ======================
class SimpleUNet(nn.Module):
    def __init__(self):
        super(SimpleUNet, self).__init__()
        self.encoder = nn.Sequential(
            nn.Conv2d(1, 16, 3, padding=1), nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Conv2d(16, 32, 3, padding=1), nn.ReLU(),
            nn.MaxPool2d(2)
        )
        self.decoder = nn.Sequential(
            nn.ConvTranspose2d(32, 16, 2, stride=2), nn.ReLU(),
            nn.ConvTranspose2d(16, 1, 2, stride=2), nn.Sigmoid()
        )
    def forward(self, x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x

lung_segmenter = SimpleUNet().to("cuda" if torch.cuda.is_available() else "cpu")
lung_segmenter.eval()

# ======================
# 2️⃣ Load DenseNet121 Model for Classification
# ======================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = models.densenet121(pretrained=True)
model.classifier = nn.Linear(model.classifier.in_features, 2)
model = model.to(device)
model.eval()  # Load your trained DenseNet weights here

# ======================
# 3️⃣ Test Dataset Loader
# ======================
test_dir = "/content/Pediatric_chest_xray/Pediatric Chest X-ray Pneumonia/test"
test_transforms = transforms.Compose([
    transforms.Grayscale(num_output_channels=1),
    transforms.Resize((224,224)),
    transforms.ToTensor(),
    transforms.Normalize([0.5],[0.5])
])
test_dataset = datasets.ImageFolder(test_dir, transform=test_transforms)
test_loader = DataLoader(test_dataset, batch_size=16, shuffle=False)

# ======================
# 4️⃣ Grad-CAM + Masked Images
# ======================
target_layer = model.features[-1]
cam = GradCAMPlusPlus(model=model, target_layers=[target_layer])

def tensor_to_cam_image(img_tensor):
    img_np = img_tensor.cpu().numpy().transpose(1,2,0)
    img_np = (img_np * 0.5 + 0.5).clip(0,1)
    # Convert grayscale to 3 channels
    if img_np.shape[2] == 1:
        img_np = np.repeat(img_np, 3, axis=2)
    return img_np

def apply_lung_mask(img_tensor):
    with torch.no_grad():
        mask = lung_segmenter(img_tensor.unsqueeze(0).to(device))
        mask = mask.squeeze().cpu().numpy()
        mask = np.expand_dims(mask, axis=2)
        img = tensor_to_cam_image(img_tensor)
        img_masked = img * mask
        return img_masked

def collect_balanced_images(test_loader, max_per_class=4):
    class_counts = {0:0, 1:0}
    selected_imgs = []
    selected_titles = []

    for images, labels in test_loader:
        for i in range(len(labels)):
            label = labels[i].item()
            if class_counts[label] >= max_per_class:
                continue

            img_tensor = images[i].unsqueeze(0).to(device)  # [1,1,224,224]
            img_tensor = img_tensor.repeat(1,3,1,1)         # Convert to [1,3,224,224] for DenseNet

            # DenseNet prediction
            pred_class = torch.argmax(model(img_tensor), dim=1).item()

            # Apply lung mask (keep 1 channel for mask, multiply later)
            img_masked = apply_lung_mask(images[i])

            # Grad-CAM
            grayscale_cam = cam(input_tensor=img_tensor, targets=[ClassifierOutputTarget(pred_class)])[0]
            cam_image = show_cam_on_image(img_masked, grayscale_cam, use_rgb=True)

            title_color = 'green' if pred_class == label else 'red'
            title_text = f"P: {test_loader.dataset.classes[pred_class]}\nT: {test_loader.dataset.classes[label]}"

            selected_imgs.append(cam_image)
            selected_titles.append((title_text, title_color))
            class_counts[label] += 1

        if all(v >= max_per_class for v in class_counts.values()):
            break

    return selected_imgs, selected_titles

def plot_grid(images, titles, cols=4, figsize=(16,12)):
    rows = (len(images)+cols-1)//cols
    fig, axes = plt.subplots(rows, cols, figsize=figsize)
    axes = axes.flatten()
    for idx, img in enumerate(images):
        axes[idx].imshow(img)
        title, color = titles[idx]
        axes[idx].set_title(title, color=color)
        axes[idx].axis('off')
    for ax in axes[len(images):]:
        ax.axis('off')
    plt.tight_layout()
    plt.show()

# ======================
# 5️⃣ Execute Grad-CAM visualization
# ======================
images, titles = collect_balanced_images(test_loader, max_per_class=4)
plot_grid(images, titles)
```
